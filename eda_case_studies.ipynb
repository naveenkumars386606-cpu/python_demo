```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("/content/vgsales (1).csv")   # change path if needed

print("Initial Shape:", df.shape)


```

```python
df.drop(columns=["Rank"], inplace=True)

print("After Dropping Rank:", df.shape)
```

```python
df["Year"].fillna(df["Year"].median(), inplace=True)

# Fill Publisher with 'Unknown'
df["Publisher"].fillna("Unknown", inplace=True)

print("Null values after handling:")
print(df.isnull().sum())
```

```python
print("Duplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", df.shape)
```

```python


sales_cols = ["NA_Sales", "EU_Sales", "JP_Sales",
              "Other_Sales", "Global_Sales"]

for col in sales_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])

print("Outliers capped successfully")
```

```python
object_cols = df.select_dtypes(include="object").columns
print("Object Columns:", object_cols)

le = LabelEncoder()

for col in object_cols:
    df[col] = le.fit_transform(df[col])

print("Encoding completed")
```

```python
print("Final Shape:", df.shape)
print(df.head())
```

```python
# STEP 0: Load Dataset
# ==============================

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("commodity_prices.csv")

print("Initial Shape:", df.shape)
print(df.isnull().sum())
```

```python
# STEP 1: Drop Unnecessary Columns
# ==============================

# Drop index-like column if present
if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)

print("After Dropping Columns:", df.shape)
```

```python
# STEP 1: Drop Unnecessary Columns
# ==============================

# Drop index-like column if present
if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)

print("After Dropping Columns:", df.shape)
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Fill numeric columns with median
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Drop rows where date is missing
df.dropna(subset=["date"], inplace=True)

print("Null values after handling:")
print(df.isnull().sum())
```

```python
print("Duplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", df.shape)
```

```python
# STEP 4: Handle Outliers (IQR Capping)
# ==============================

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])

print("Outliers handled using IQR capping.")
```

```python
# STEP 5: Encode Object Columns
# ==============================

object_cols = df.select_dtypes(include="object").columns
print("Object Columns:", object_cols)

le = LabelEncoder()

for col in object_cols:
    df[col] = le.fit_transform(df[col])

print("Encoding completed.")
```

```python
print("Final Shape:", df.shape)
print(df.head())
```
