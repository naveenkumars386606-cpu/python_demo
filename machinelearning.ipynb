```python
import pandas as pd

df=pd.read_csv("/content/bank-additional-full-1.csv",sep=";")
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
for i in df.columns:
  if(df[i].dtype != "object"):
    df[i] = df[i].fillna(df[i].mean())

  else:
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in df.columns:
  if(df[i].dtype !="object"):
    plt.boxplot(df[i])
    plt.title(i)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in df.columns:
  if(df[i].dtype == "object"):
    df[i] = le.fit_transform(df[i])
```

```python
df.info()
```

```python
# 1. Split the data into x,y
# 2. Split the data into train and test
# 3. Train the model
# 4. Test the model
# 5. Evaluate the Model
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
```
